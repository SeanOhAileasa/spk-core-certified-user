###### ``| sort``
***

The ``sort`` command allows you to display a field's values in ascending or descending order. 

The syntax for the sort command is as follows: 

- ``| sort`` followed by a minus or plus sign - will determine whether you're sorting in ascending order by default with the plus sign or in descending order with the minus sign. The plus sign is implied, so you don't have to use it deliberately. 


- Follow this with the field or fields you wish to execute the ``sort`` based on. If the field name on which you're sorting contains spaces or non-alphanumeric characters, be sure to enclose it in double quotes. 


- Finally, if you only want a certain number of results, you can use the ``limit`` equals argument, or just put an integer representing the number of results that you want. 

Splunk will then determine the data type present in your field and sort appropriately. Alphabetic strings will be sorted lexicographically, numbers will be sorted numerically, and so on. 

Note that the sorting of alphabetic strings is done lexicographically as opposed to alphabetically.

This means that uppercase letters appear before lowercase. We can see this in the example shown here, where ``Bee`` with a capital ``B`` gets placed before ``ant`` with a lowercase ``a``. 

- Asked to display vendor information for vendors that begin with "Bea":

We have our table already set up - now just need to sort.

- Sort our rows based on descending order in the Vendor field:

Note that we have a space between the minus sign and the field that we're sorting by. 

This space can become important if you want to sort on more than one field.

Let's take a look at the difference in two examples.

- First, let's add a comma and then another field to sort by - ``VendorCity``:

Since we have a space between the minus sign and the first field in our list, the descending order instruction is applied to all fields in the list. 

Can see this in the ``VendorCity`` values listed for the ``Beach Games`` vendor, ``Miami`` comes before ``Fort Lauderdale`` because descending order is applied in both cases.

Here, ``Beach Games`` still comes last in the ``Vendor`` column, but now ``Fort Lauderdale`` comes before ``Miami`` in the ``VendorCity`` column.
