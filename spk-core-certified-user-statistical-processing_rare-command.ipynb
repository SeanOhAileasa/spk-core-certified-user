###### ``| rare``
***

The ``rare`` command has the same options as the ``top`` command, but it returns the least common values of a result set. 

Essentially, the ``rare`` command is the opposite of the ``top`` command. 

Here we are looking for the products that have the least number of sales by Buttercup Games vendors in a given time period - note that the same arguments and options available in the ``top`` command are also available with ``rare``.
