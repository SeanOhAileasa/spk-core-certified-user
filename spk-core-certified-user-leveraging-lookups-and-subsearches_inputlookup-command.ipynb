###### ``| inputlookup``
***

The inputlookup command can be useful for searching and validating the contents of a lookup table. 

The inputlookup command is a generating command, which means that it is going to be the first command in our search, and it will not be preceded by a search. 

We will follow the inputlookup command with the name of the file or the lookup definition that we would like to examine and validate. 

We are passing in the lookup table file to the inputlookup command, to verify whether the contents of this ``products.csv`` file were loaded in properly.

Will display the set of column names and the values down each column that exist within that CSV.
