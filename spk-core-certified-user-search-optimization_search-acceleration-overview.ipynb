###### Search Acceleration Overview
***

While Splunk can scan massive amounts of data, the time it takes to return the results will be affected by the number of events summarized in the report. 

In cases where you're running a one-off search, this might not matter much, but it quickly can become impractical when you are using these searches in frequently run reports and dashboards, taxing your deployment and possibly slowing down search result times for all users. 

Splunk allows you to create summaries of your event data - these are smaller segments of event data populated by background searches that only include the data needed to fulfill the search. 

When you run a search against one of these summaries, it will complete significantly faster since the data you're searching over is much smaller than the original raw events.

> Data Summary Creation Methods

Splunk has three ``Data Summary Creation Methods``: 

- Report Acceleration


- Summary Indexing 


- Data Model Acceleration

> ``Report Acceleration`` 

Uses automatically created summaries to speed up completion times for certain kinds of reports. 

> ``Summary Indexing``

Enables search acceleration through the manual creation of summary indexes. 

These indexes exist separately from other indexes in your deployment. 

> ``Data Model Acceleration``

Data model acceleration will accelerate all fields defined in a data model. 

Data model acceleration is the easiest and most efficient option and will typically be your first choice. 

***
###### Notes
***

[ [``Summary Indexing``](https://docs.splunk.com/Documentation/Splunk/latest/Knowledge/Usesummaryindexing) ] is recommended only for reports that don't qualify for acceleration.
