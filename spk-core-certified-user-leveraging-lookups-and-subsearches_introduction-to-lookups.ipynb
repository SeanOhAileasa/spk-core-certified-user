###### Introduction to Lookups
***

Cover:

- ``| lookup``


- correlate events with subsearches


- ``| return``

Lookups allow you to add other fields and values to your events not included in your indexed data. 

We can combine fields from sources external to the index with searched events, based on paired fields present in the events. This might include CSV files, scripts, or geospatial data.

Some examples would be tying a user's personal information to their RFID, or text descriptions for product IDs. Data that would be useful for search, but might not be available in your index. 

After a lookup is configured, you can use its fields in search, and it shows up in the fields sidebar.

By default, there are four types of lookups. 

- File-based lookups populate your events with fields pulled from CSV files. 


- External lookups append data to your events through the use of Python scripts or binary executables. 


- KV Store lookups access key-value pairs from a KV Store collection, and geospatial lookups. 


- Geographic information to your events by referencing a KMZ or KML file.
