###### Eval as a Function
***

The ``eval`` command can also be used as a function with the ``stats`` command. 

Been asked to count the number of events that occurred in a given time period where the ``vendor_action`` was ``Accepted``, ``Failed``, or ``session opened``. 

Using ``eval`` as a function within the ``count`` function. 

The search is as follows:

- Giving a statistical analysis (``stats``) and ``count`` events where now we ``eval`` where the value of ``Accepted`` is found in the field of ``vendor_action``, relabel that a field called ``Accepted``.


- Further ``count`` where the value of ``Failed`` is found in the field of ``vendor_action``, call this a new field called ``Failed```.


- Finally ``count`` all the events where a value of ``session opened`` is found in the field of ``vendor_action``, and call this a new field called ``SessionOpened``. 

This allows us to arbitrarily group values together that would otherwise be seen as multiple different splits - does require an ``as`` clause to rename the newly created field.
