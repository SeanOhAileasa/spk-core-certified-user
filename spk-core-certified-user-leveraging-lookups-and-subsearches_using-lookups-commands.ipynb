###### Using Lookups Commands
***

Taking a look at some commands that can be used to interact with lookups in searches. 

Lookups are a way to add additional fields to our search results at search time. 

As we can see in this example, we are looking at some badge reader data and if we expand an event, we can see the fields associated with that specific event. 

There is an ``rfid`` number, which stands for Radio Frequency ID Number, that is associated with an address and address description, but we don't necessarily have any information about who the user, first name, last name, email, and department information is. 

So in that case, we are going to leverage a lookup by bringing in the set of fields that are associated with each specific ``rfid`` value, and look at the associated user with the details of that user.
