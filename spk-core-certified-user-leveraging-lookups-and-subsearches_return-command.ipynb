###### Return Command
***

The return command is used to pass values from a subsearch to the outer search. 

In most cases, using the return command at the end of a subsearch removes the need for the fields command, a rename command, format, dedup, or head command. 

The return command is preceded by a search and followed by a count value, which is considered an integer, which will specify and tell Splunk how many rows of results to return - by default, it will only return the first row. 


From there, we can provide one or more fields to return, separated by spaces.

We have the option to just return the values of the field, or provide a new alias to rename the field.

- In this first search, we are returning src_ip, which will by default return the first field-value pair with the src_ip field and its value. 

- If we place a dollar sign in front of that src_ip field, it will only return that IP address value and exclude the field name from being displayed in the results. 

- If we provide an alias in front of the equal sign, followed by the name of the field that we're working with, in this case "src_ip", this will rename src_ip to that aliased field name of "ip". Typically this is done if the outer search results have a field named differently than the field name in the inner search results. In this case, the field may be named "ip" and we can quickly rename it using the return command.

If we're comparing return to the fields command:

- Can see here that the fields command will display and return all of the field-value pairs of src_ip. 

- If we want a little bit more control in terms of how many results are sent to the outer search, in this case, let's say the first three results, we can include the return command with a limit of 3, followed by src_ip.
