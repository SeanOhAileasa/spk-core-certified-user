###### Search Optimization Overview
***

Cover how to optimize a search, accelerate reports and data models, and about some commands you can use to best access those data models - starting with an overview of the basics. 

> ``Search Modes``

Determine how much field data is returned, and as a result, affect how fast the search completes. 

There are three search modes to choose from: 

- ``Fast Mode``


- ``Smart Mode``


- ``Verbose Mode``

> ``Fast Mode``

Prioritizes speed over completeness. 

**Field discovery is disabled for this mode, only returning information on default fields and fields required to fulfill your search.**

> ``Verbose Mode``

Prioritizes completeness over speed, returning all extracted fields, an event list, and in the timeline for every search. 

Due to the increased size of the search payload, verbose mode is the slowest search mode option. 

> ``Smart Mode``

Is the default search mode, balancing speed and completeness. 

> Transforming Search

If a transforming search is run, it will behave like fast mode. 

If a non-transforming search is run, it behaves like verbose mode.

> ``_time``

Using ``_time`` to limit the events returned is the most efficient way to filter events - the less data you have to search, the faster Splunk will be. 

> ``index`` - ``source`` - ``host`` - ``sourcetype``

After time, the default fields of index, source, host and sourcetype are the most powerful.

These fields are extracted at index time, so they will not need to be extracted for each search. 

> Best Practices

The more you tell the search engine, the more likely it is that you'll get good results. 

- It's better to search for "failed password" instead of just "password". 


- Inclusion is generally better than exclusion. 


- Searching for ``"access denied"`` is better than searching for ``NOT "access granted"``. 


- When possible, use the OR or IN operators instead of wildcards.

```
"user=admin OR user=administrator"
```

```sql
user IN (admin,administrator)
```

Apply filtering commands as early as possible in your search. 

Filtering early limits the number of events returned, making future manipulations of the data faster.
