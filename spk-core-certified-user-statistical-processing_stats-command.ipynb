###### ``| stats``
***

To produce statistics from our search results, we can use the ``stats`` command. 

The ``stats`` command uses various functions to calculate statistics on search results - output in this case will be a table. 

Can also choose to group your results with a ``by`` clause. 

Unlike ``chart`` and ``timechart``, ``stats`` allows for continuous splits of your data - can group more than two fields together with the ``by`` clause here. 

The ``wc-field`` means that the ``stats`` command supports wildcard fields. 

Like ``chart`` and ``timechart``, ``stats`` supports statistical functions - but unlike ``chart`` and ``timechart``, ``stats`` can support multiple fields. 

Are asked to count the invalid or failed login attempts during a given time period - note that we can rename the resulting ``count`` column with an ``as`` clause. 

Do not choose to rename a field with the ``as`` clause then will be named according to the search syntax. 

In this example, we can see both a count of events that contain a value in the ``vendor_action`` field, as well as the overall total of events returned in our base search. 

This slide further demonstrates how more than one statistical function can be invoked with the ``stats`` command.

Can do this by separating the functions with commas - can also see here how the ``as`` clause allows us to rename fields inline in our search. 

- Counting the total number of events that contain a value in the ``vendor_action`` field and renaming that result as ``ActionEvents``. 


- Further counting up the total number of events returned to the base search and calling this ``TotalEvents``.

**In the event that we use multiple functions, it's a very good idea to make sure that we rename our results in order to eliminate confusion.**

As with ``chart`` and ``timechart``, the ``by`` clause allows us to split our resulting table for ``stats`` based on the values in a given field or fields. 

The order in which the fields are listed has a big impact on the results - your data will first be grouped by the first field given, then grouped by the second field given and so on.

First grouped our returned results by the values found in the ``user`` field and then by the values found in the ``tag`` field within each user. 

- Switching the ``tag`` and ``user`` values in the search:

In this table, we can see our returned results are first grouped by values found in ``tag`` and then grouped by values found in ``user``.
