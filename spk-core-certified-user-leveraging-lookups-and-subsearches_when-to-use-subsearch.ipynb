###### When to Use Subsearch
***

Subsearches are useful, but they have their limits - multiple subsearches can be used in a search, and subsearches can be nested.

**Careful: Each subsearch requires an additional search on your deployment, which could quickly eat up memory and CPU.**

They have a default time limit of 60 seconds to execute, after which they are finalized - once that time limit is reached, only the events found during those 60 seconds are returned to the outer search, whether the search is complete or not. 

Subsearches also have a results limit - by default, they only return up to 10,000 results - once it's met, search stops, resulting in a partial set of events. 

**Subsearch results limits can be adjusted only by a Splunk administrator.**

Because of these limitations, we should focus our subsearch on a narrow set of data - iideally, a subsearch produces a small set of results filtering out unnecessary events.

It then returns that narrow set of results for the outer search to work with. 

If an outer search is set to execute in real-time, then its subsearch will execute over all time by default. 

In order to avoid exceeding the subsearch's time limit, we can include ``earliest`` and ``latest`` time modifiers in the subsearch.

Subsearches are slow - after all, there are two searches instead of one. 

**If you can use a combination of ``stats`` and ``eval`` in a single search instead, they typically perform better.**

**A rule of thumb should always be to only use a subsearch when there is no other way to retrieve the data you need.**
