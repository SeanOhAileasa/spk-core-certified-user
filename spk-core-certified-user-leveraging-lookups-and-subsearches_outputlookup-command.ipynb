### Leveraging Lookups and Subsearches
***

< [Table of Contents](#top) | [References](#references) >
<a id="LeveragingLookupsandSubsearchese"></a>
***
###### ``| Outputlookup``
***

To write the results of a search to a lookup, you can use the outputlookup command. 

The outputlookup command is preceded by a search. 

That search's results are going to be written to the specified ``<filename>`` or ``<lookup-definition>`` that is specified in the outputlookup command.

Can see an outputlookup command writing those search results to a CSV file named ``users.csv``, or to a lookup definition, table name of ``usergroup``. 

That lookup will be created in the app that we are currently running that search in, by default. 

Otherwise if we set ``createinapp=false`` which is an option built in to the outputlookup command, that is going to save that lookup file within the system lookups directory. 

One example of using the outputlookup command is running a search that is saved as an alert, where we are using it to create a searchable lookup. 

- Searching across our failed linux_secure events across the last 30 days. 

Going to get a ``count`` of events by ``user``.

Compute the daily average by dividing the ``count`` by 30 and ``round`` it to the nearest integer level. 

From there, writing those results over to the CSV file called ``averages.csv``, which is going to be created in the application that we're currently working in. 

From there, we can save this as a report or an alert, run it on a schedule, and that ``averages.csv`` file will be updated frequently.

Can then use the inputlookup command to verify the contents of it.
