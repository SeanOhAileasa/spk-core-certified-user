###### Formatting Time
***

Can define how time is formatted in the search results by using the time functions of the ``eval`` command. 

> ``| eval``

The eval command is a very powerful command that allows us to write the results of an expression to a new or preexisting field. 

We can chain multiple eval expressions together by comma-separating them and the eval command supports its own set of functions.

> ``| eval`` - ``now()`` - ``time()``

Some of those functions being date and time functions that allow us to generate time values and manipulate time.

If we take a look at the ``now`` function, the now function returns the time that a search was started.

The ``time`` function returns the time that an event was processed by the eval command. 

We can set those functions as values of the fields that we're creating with the eval command that will store those particular timestamps. 

> ``| eval`` - ``relative_time()``

The ``relative_time`` function is a function that takes in two arguments:

- the first argument being the actual time value - that time value will be represented in epoch time 


- the second argument being a specific time specifier using the time unit abbreviations

Creating a new field called yesterday, setting it equal to the ``relative_time`` function. 

The first argument will be the time generated by the ``now`` function. 

The second argument will be based on the time right now - going to look back one day ago, snapping to the beginning of the current hour. 

What this is going to result in is a time value which is represented in epoch time, stored as a value for that field of yesterday. 

> ``| eval`` - ``strftime()`` - ``strptime()``

For working with the ``strftime`` and ``strptime`` functions of the eval command: these have date time unit abbreviations, each one representing a different unit of time.

- ``Splunk > docs`` [ [``Date and time format variables``](https://docs.splunk.com/Documentation/Splunk/9.0.2/SearchReference/Commontimeformatvariables) ]

The epoch time generated by this relative_time function, which is stored in the yesterday field, is going to be formatted into a more readable format by this strftime function, where the time value that's stored in yesterday will be passed in as the first argument - second argument being in the format of year, month, day, hour and minute:

- Run a new search using this strftime function where we are formatting the time values. 

If we take a look at our retail sales data, ``index`` equal to ``sales``, ``sourcetype`` equal to ``vendor_sales``, and we display our data in a ``timechart`` where we're interested in looking at the hourly sales, and we would like these time buckets to each be one hour apart:

We can see the ``_time`` column contains these time values displayed in the month, day, year, hour, minute format. 

Would like to format this into something a little more readable using the strftime function, so will pipe to the eval command, and update ``_time``. 

As an alternative, could create a new field just to compare it to ``_time``, so we can create a new field name, let's say, ``formatted_time``, and set that equal to the strftime function with ``_time`` as the first argument - then we're going to use these date time format variables:

- ``%b`` being the abbreviated month


- ``%d`` being the day of the month


- ``%I`` being the hour


- ``%p`` being AM or PM

Run the search, and we can see ``formatted_time`` being displayed.

The difference between the strftime and the strptime function is that the ``strptime``  is going to take a time represented by a string to a Unix timestamp, so we're essentially going from a formatted value to epoch time.

- First argument will be the field that contains those formatted time values (``asctime``). 


- Second argument using the date time format variables in the order that the format is specified in for that first argument - ``asctime``. 

This will result in a newly created field with those epoch time values.
