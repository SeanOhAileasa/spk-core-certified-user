###### Splunk Search Scheduler
***

The Splunk Search Scheduler allows you to define and manage when scheduled reports and alerts will run. 

There's a limit to the number of searches Splunk can run concurrently, generally based on the number of CPUs, which is tuneable.

When Splunk is running too many searches, the next search that Splunk chooses to run is based on a priority: 

- When the slot becomes available for Splunk to run a search, the first choice will be ad hoc historical searches. 

These are searches that are run manually by users, usually through Splunk Web. 

- After ad hoc historical searches, the next searches considered to run will be manually scheduled searches using real-time scheduling. 

Don't get confused by the term "real-time" here - not talking about real-time searches which run continuously. In this case, the phrase "real-time" is referring to a scheduling mechanism. With real-time scheduling, searches actually have the ability to be skipped. 

- After real-time searches, Splunk will choose to run manually scheduled searches with continuous scheduling. 

Continuous scheduling differs from real-time scheduling in that searches aren't actually skipped, but they can be paused, and for extensive periods of time. 

- The last consideration is automatically scheduled searches.

These are searches that were put on the schedule by automatic processes that maintain things like acceleration summaries and auto summarization.
