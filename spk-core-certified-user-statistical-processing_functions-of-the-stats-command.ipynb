###### Functions of the Stats Command
***

Previously, we used the ``count`` function to count the number of events that match specified search criteria - now taking a look at some of the other functions that are commonly used with the ``stats`` command.

Used two of the aggregate functions, ``count`` and ``sum``, in our previous examples with ``chart`` and ``timechart`` but there are many different functions that can be used with these commands. 

Broadly, these functions fall into one of four categories:

- Either allow us to **Aggregate** values.


- Sort on a dependent order (**Event Order**) of values.


- Return multiple values (**Multivalue**) from within a given field.


- Return values based on **Time**.

A small sample of some of the functions that can be used with the ``stats`` command.

> ``dc()`` - ``distinct_count()``

Provides a count of the unique values for a given field in the result set. 

Question was asked: 

- How many unique websites have employees visited in the last four hours? 

Using ``distinct_count`` (``dc``), we can get the unique values found within a given field. 

> ``sum()``

The ``sum`` function calculates the sum of values within a given field, as long as the values are numeric. 

In this example, we're being asked:

- How much bandwidth employees consumed at each website during the past week?

First set our time range to the last seven days. 

- Using the ``sum`` function to begin calculating our bandwidth - ``stats sum`` of the values found in the ``sc_bytes`` field - then split this by the values found in the ``s_hostname`` field):

Can see the different websites visited on the left, and a sum of the values in the ``sc_bytes`` field over the last seven days on the right.

- Making our results a bit cleaner for our end user by renaming some of these values - can rename ``sum(sc_bytes)`` as ``Bandwidth``:

Make it more obvious what we're measuring.

- Can ``sort`` on the ``Bandwidth`` field from highest to lowest:

Using a minus sign in front of the field you wish to sort tells Splunk that you wish to sort in descending order, i.e. from highest to lowest.

Can combine multiple functions within our ``stats`` command, as well as multiple commands within a search. 

An example where we are asked to report the number of retail units sold and sales revenue for each product during the previous week. 

- First count the number of events that have a value in the ``price`` field:

- Call this field ``Units Sold`` - within the same ``stats`` command, we should also get a sum of all the values in the ``price`` field and rename this result ``Total Sales``:  

- Splitting our results by the values found in the ``product_name`` field, thereby giving us the number of units and the total amount of money made per product:

- Can ``sort`` our returned rows based on the values found in the ``Total Sales`` field: 

Again, the minus sign lets Splunk know that we wish to have this sort done in descending order. 

> ``min()`` - ``max()`` - ``avg()``

Three more common functions that are used with the ``stats`` command are ``min``, ``max``, and ``average``. 

- ``Min`` will give you the lowest value in a given field. 


- ``Max`` will give you the highest value in a given field. 


- ``Avg`` will give you the average value in a given field. 

Can see here we are asked what is the minimum, maximum, and average bandwidth used for each website usage type? 

The ``avg`` function ignores any event or row that does not have a value for the specified field or has a non-numerical value for that field. 

> ``values()``

The values function allows us to list the unique values found in a given field. 

Showing all unique ``values`` found in the ``user`` field and labeling this as ``User Names``. 

Then ``count`` all the instances that a value is found in the user field, unique or otherwise, and label this as ``Attempts``. 

Then split our result set by the values found in the ``src_ip`` field. 

The resulting table shows us for each IP address what usernames can be found and how many times any username was present for that IP address. 

> ``list()``

The ``list`` function lists all values found in a given field, including duplicates. 

Been asked which websites has each employee accessed during the last 60 minutes - the ``list`` function can be quite useful for creating easy-to-read data tables. 

Note that if an event contains multiple values for a given field, that's fine - all values will be shown.

> ``list`` V ``values``

Let's highlight the difference between the ``list`` and ``values`` functions. 

On the left, we can see the ``list`` function being used - every instance of a value within the ``s_hostname`` field is being displayed, even if they are duplicates within a given username. 

On the right, we can see that the ``values`` function will only display unique values for a field within a given username. 

Putting it all together, we can see how taking advantage of multiple functions within a single ``stats`` command allows us to pull some very powerful insights out of our data. 

Working with a simple result set from our base search, we are still able to set up a table that explores multiple aspects of our data. 

Here we are only using values found in the ``price`` field and splitting that by the values found in the ``VendorCountry`` field. 

Even so, those two fields alone can tell us the amount of products sold, the average price it was sold for, and the range of prices from their products, among other insights
