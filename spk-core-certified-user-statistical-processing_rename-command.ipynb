###### Rename Command
***

There may be times when you want to display a more meaningful or user-friendly name for a field - to do this, we use the ``rename`` command. 

Renaming a field is fairly simple - can use the command ``| rename``, then the name of the field as it currently exists, followed by the ``as`` clause, followed by the new name we wish to give to the field. 

If your new field name contains special characters or spaces, be sure to use double quotes to encompass the entirety of your new field name. 

Renaming multiple fields within the same ``rename`` command. 

First chosen to rename the ``productId`` field as ``ProductID`` with a capital P and capital ID, followed by a comma, then we ``rename`` the ``action`` field as ``Customer Action``, and finally, we ``rename`` the ``status`` field as ``HTTP Status``. 

The slide also points out a nice example that your clauses such as ``as`` or ``by`` do not need to be upper or lower case - these are in fact case insensitive. 

Note also that as "Customer Action" and "HTTP Status" both contain spaces in their field names, we need to encompass the entirety of these field names with double quotes. 

Once a field is renamed, Splunk will only respond to or be able to use the new name of that field for the rest of your search. 

The example on the left shows what can go wrong. In this particular case, the user has renamed the ``action`` field as ``Customer Action`` and the ``status`` field as ``HTTP Status``. They then asked to get a table with the fields ``action`` and ``status``. However, since these fields were renamed, Splunk no longer recognizes that there was ever a field called ``action`` or ``status``. 

On the right hand side, we see a better example. In this case, the user has further called a new command, ``sort``, as opposed to ``table``. But this user has made sure to use the new names given to each field. As such, Splunk has no problem creating this table. 

Can also use wildcards to ``rename`` multiple fields to match a pattern. 

In this example, we've renamed the fields ``productId`` and ``product_name`` by changing ``product*`` to ``PROD``. When we then base a table on ``PROD``, we can see that both of our original fields will appear.
