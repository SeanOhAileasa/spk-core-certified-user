###### ``| top``
***

The ``top`` command finds the most common values for given fields in a result set. 

Taking a look at the syntax - the command starts with ``| top``, followed by a list of fields we want to find the most common values for. 

Can also group results together based on shared field with the ``by`` clause. 

The ``top`` command will return a table with the top 10 most common results by default, but this can be overridden with the ``limit`` argument. 

The optional arguments that can be used with the ``top`` command - ``count``, ``field``, ``limit``, and ``showperc``. 

In this scenario, we are asked to determine which IP addresses generated the most attacks in the last 60 minutes.

Without any additional arguments, we can see that this search command returns a table which lists the top 10 most common IP addresses returned in our base search - can see that a percentage (``percent``) is also included - this is because ``showperc`` will default to ``true``.

Here we can see the ``limit`` argument being used - by adding the argument ``limit=5`` to our ``top`` command, we limit the return results for the ``src_ip`` field to just the top five most common. 

This search asks us to display the top five IPs that generated the most attacks during the last hour.

- Another important aspect of the ``top`` command is the ability to find top values for specific combinations of fields by adding more than one field to our field list:

Here we can see that our table is showing results from the top 10 most common values in the ``cs_username`` field.

In our previous search, all values for the ``cs_username`` field were unique - in this search, however, all combinations of ``cs_username`` and webcat code are unique. 

The ``top`` command will return both a ``count`` and a ``percent`` field by default.

Can choose to toggle the percentage field - with the boolean argument, ``showperc`` to ``false``, ``f``, or ``0`` will prevent the ``percent`` column from being displayed in the resulting table. 

``Countfield`` is another argument that can be added to the ``top`` command. 

In this case, the ``countfield`` argument allows us to rename the ``count`` column from the default of ``count`` to whatever string we choose - in this example, we can see that the ``count`` column has been renamed ``Total Viewed``.
