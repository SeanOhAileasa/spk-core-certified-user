###### What is a Data Series?
***

Cover the commands and functions that can help you transform, manipulate and format your data. 

A ``Data Series`` is a sequence of related data points that are plotted in a visualization. 

Three types of ``Data Series``: 

- ``Single Series``

Data is a sequence of related data points that compares values of a single data category.

- ``Multi-series``

Multi-series data is a sequence of related data points that compares values of two or more data categories.

- ``Time Series``

Time Series data is a sequence of related data points that compares values over time, which can be either Single or Multi-series.

> Transforming Commands

Can be used in searches to organize your results into a statistical table containing a data series that can be visualized.
