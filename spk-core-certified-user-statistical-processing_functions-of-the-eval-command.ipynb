###### Functions of the Eval Command
***

``eval`` functions allow you to convert, compare and perform mathematical calculations on values in your data.

Taking a high level look at some of the most commonly used eval functions. 

Evaluation functions will evaluate an expression based on your events and then return to your result.

Now, this is not necessarily true for all functions in ``eval`` - some do not evaluate any expressions, but instead return a result based on their own functionality. 

There are 11 different categories of evaluation functions - some will convert values, some will compare values, some will perform mathematical operations, some will perform statistical operations, others will transform text, and many more.

A small sample of some of the functions available with the ``eval`` statement. 

- ``round`` will round X to Y decimal places - otherwise, if not given an argument of Y, we return X as a whole number. 


- ``pow`` will return X raised to the power of Y. 

Also have some statistical functions, such as ``min``, ``max``, and ``random``. 

> ``pow()``

In this particular case, we've created a new field called ``bandwidth``. 

The expression that informs us of the values of ``bandwidth`` is the following:

- values in the field of ``Bytes`` divided by 1024 raised to the power of two

In this example, we can see that ``pow`` is given two arguments

- ``X`` is the number we wish to raise: 1024


- ``Y``, is the number of powers we wish to raise it to

> ``round()``

The ``round`` function is also extremely commonly used. 

We will take X and round it to the decimal places specified by Y. 

In the example shown below, on the right hand side, we can see the field of bandwidth has many decimal places - probably more than we need. 

By calling ``eval bandwidth = round`` - passing ``round`` the arguments of ``bandwidth`` and ``2``, we can see that we have rounded bandwidth to two decimal places. 

It's a common practice to wrap the ``round`` function around the results of another statistical function. 

This way, we make sure that our end result of the entire process is human-readable. 

> ``max`` - ``min`` - ``random``

> ``random`` Example

Here we see a slight variation of the classic dice roll example of ``random``. 

Creating a new field called ``group``, and the values in ``group`` would be as follows: 

- the output of the ``random`` function to the ``%5``


- then add ``1`` to make sure that our result cannot be zero - will give us a random number between ``1`` and ``5``. 


- then choosing to ``list`` the values in the ``user`` field as ``Users`` by ``group`` - would allow us to create five random groups of users. 

> ``eval`` Without Functions

The ``eval`` expression can be used without functions. 

It doesn't necessarily require a function - we can see the creation of a field called ``Discount``. 

The expression that informs us of the values in the ``Discount`` field are ``tlp - tsp``, divided by ``tlp``, no function necessary. 

Later on in the same search, we can see ``eval`` and ``Discount`` again and in this case, we're simply concatenating (``.``) the current value in ``Discount`` with a percentage sign.
