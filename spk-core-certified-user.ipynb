### What is Splunk
***

> ``Machine Data`` / ``Operational Intelligence``

``Machine Data`` is everywhere, constantly flowing from the devices we interact with - from the car we take to work, the buildings we enter, the servers we connect to, even the phone in our pocket, making up for more than 90% of the data accumulated by today's organisations. 

This data can be in many different formats, some of its structured, some of it not, making it unpredictable and difficult to process. 

> Scenario

Taking a look at a scenario from a fictional gaming company with offices across the globe. 

A customer calls support to report having an issue (purchasing in game upgrades for their product). The support console showed that everything should be working correctly. The issue was escalated to application support - checked all monitoring tools and came up empty handed. 

So the issue was escalated to the developer, to look at the code and run some tests - not seeing any issues, got the system administrator involved - sifted through the application, web server and point of sale server logs. 

It was established it was not an application error, so it was escalated to the database administrator - started going through the SQL logs, found a slow query making it hard for some users to complete their transactions. 

At this point, nearly nine hours had passed and while the issue was found, it still needed to be fixed and that was just one issue that popped up that day. 

Used ``Machine Data`` to solve this issue in less time: 

- When the initial support call comes in, it is logged for future use.

Taking it one step further - what if we could alert when transactions fail even before the customer calls tech support?

This is where ``Splunk`` comes in - can pinpoint, correlate and alert on specific events (could save our fellow employees many hours and headaches).

Simply put ``Splunk`` can take any data and add it to an intelligent, searchable index, adding structure to unstructured data, and allowing you to extract all sorts of insights into your business.

Not just application issues, but security, user behaviour, hardware monitoring, sales totals (whatever you can think of doing):

> ``Splunk``

Harnessing ``Machine Data`` can give you powerful business insights. 

``Splunk Enterprise`` is an on-premises solution that allows you to ``Search``, ``Analyse`` and ``Visualise`` ``Machine Data``. 

``Splunk Cloud`` provides scalable access to the tools of ``Splunk Enterprise`` with minimal infrastructure requirements.

``Splunks`` primary features make ``Machine Data`` available, accessible and usable to everyone in your organisation. 

At the heart of ``Splunk`` is the ``Index`` which contains your ``Machine Data`` from sources such as servers, network devices, and web applications:
