###### Transforming Commands Summary
***

The ``chart``, ``timechart``, and ``stats`` commands share similar features, but it's important to use the proper command to get the results that you're looking for. 

This chart describing the differences in usable arguments across all three commands. 

First and most important is the number of arguments you can split your results by. 

- When using the ``by`` clause with ``chart``, you are limited to two fields you can base your split clause on. With ``timechart``, you only get one, that's because ``_time`` is automatically selected for you as your primary split. ``stats`` has no such limitation. 

Also of interest are some of the arguments available to you with ``chart`` and ``timechart`` that are not available with ``stats``. 

- ``limit`` limits the number of visual series shown in a **Multi-series** split. 


- Filtering out NULL (``usenull``) and other (``useother``) also refer to visual elements that you may wish to remove. 


- An X-axis also only is present when referring to some sort of visual chart. 

**The rule of thumb to remember is that the end product of the ``stats`` command is meant to be a table, while the end product of the ``chart`` and ``timechart`` command is likely a visualization.**
