###### Working with Time Zones
***

If you've ever worked with time-based data for multiple geographic locations, you know time zones can be tricky.

In this segment, we'll learn how time zones are represented in your data and use the ``strftime`` function to correct time zones in search results. 

One thing to keep in mind is to always check your data to verify that it is returning the expected results. 

What we see here as the results of this search that contains a ``bin`` command as well as the ``date_hour`` fields displaying the early morning retail sales for two to five a.m. across the previous two days.

We can see that we are displaying nine to 11 p.m., whereas we're expecting two to five a.m.

One thing to keep in mind is your time zone. 

Remember that the ``date_`` time fields do not take into account the time zone of Splunk Web, so in this case, we want to make sure that we are setting our time zone to the default system time.

Or as an alternative, using the ``strftime`` function to organize our data and normalize it. 

``strftime`` function - can pass in the ``_time`` field as its first argument, second argument will be the specific hour (``%H``). 

The difference here between the ``date_hour`` and ``my_hour``:

- **``my_hour`` is taking into account the time zone that is set**

Using the ``eval`` command with the ``strftime`` function uses the timestamp that is aligned with your time zone to retrieve the hour from the event and return a human readable format.

The ``date_hour`` field will not adjust the hour based on your Splunk Web time zone.

Rearranging the search from earlier with the ``strftime`` function, as opposed to using the ``date_hour`` fields. 

If you'd like to use your set time zone, we can create a new field called ``my_hour``, equal to the ``strftime`` function where we are formatting ``_time`` to be displayed with that preserved time zone.
