###### Formatting Visualizations
***

Explore some additional formatting options in Splunk, beginning with statistical tables. 

After running a search using a transforming command, we see a ``Format`` link in the ``Statistics`` tab:

We have options to ``Wrap Results``- show ``Row Numbers`` - change the ``Click Selection`` from cell to row - add a ``Data Overlay``.

The ``Data Overlay`` can be a heat map of values:

Or highlight the high and low values in the table:

The ``Summary`` tab allows us to add ``Totals`` and ``Percentages`` to columns.

When selecting ``Totals`` - numerical totals for a column are added up and displayed in a new row at the bottom of the table. 

The ``Percentages`` options will compute the percentage for each column and create a new row at the bottom of the table.

The ``Chart Overlay`` setting comes in very handy when using ``trendline``. 

- This column chart displaying our web sales data uses the ``timechart`` command to get a sum of all products sold over the last seven days, then pipes to the ``trendline`` command to see the trend of those sales:

We open the ``Format`` menu, select ``Chart Overlay``, and have options to either ``Overlay`` our **sales** or **trend** values:

Selecting ``trend`` overlays a line graph of ``trend`` in our data on top of the column chart displaying ``sales``.
