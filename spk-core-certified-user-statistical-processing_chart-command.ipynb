###### ``| chart``
***

The ``chart`` command is used to take results and return them formatted into a table. 

This ``chart`` command, which will return results in a table format that can be displayed as a visualization. 

- The ``| chart`` section calls the command.


- Then followed by the particular statistical function ``<stats-func>`` we are interested in using - for example, ``count`` or ``sum``.


- ``<wc-field>`` in this example would be a field that contains numeric values.


- Then can specify the X-axis of our visualization and also the first column in our resulting table with the ``over`` clause followed by the field we wish to split our rows (``<row-split>``) by. 


- Splits of our columns (``column-split``) resulting in **Multi-series** data series is accomplished by using the ``by`` clause. 

Also have the option to control our results further with the ``span``, ``limit``, ``useother`` and ``usenull`` arguments. 

Splitting our results based on the values and a set of fields is accomplished with the ``over`` and ``by`` clauses. 

- Create a ``Single Series`` split with the ``over`` clause 


- Create a ``Multi-series`` split by adding a ``by`` clause. 

Unlike the ``stats`` command, the ``chart`` command can only be split over two fields or dimensions - ``chart`` command with two arguments after the ``by`` clause is equivalent to using an ``over`` and ``by`` clause - these are functionally identical. 

> ``| chart`` - Split - **Single Series**

- Initial search produced a number of results:

- Can then choose to count these results with the ``chart count`` section of the search:

- Can then choose to split that count based on how many events contain a given value in a particular field - splitting our data by the ``VendorCountry`` field:

This result can either be displayed as a table on the ``Statistics`` tab.

Or in a graphical representation on the visualization tab.

> ``| chart`` - Split - **Multi-series**

A **Multi-series** split, one that incorporates both the ``over`` and ``by`` clauses. 

Keeping our X-axis and the first column in our chart as the ``VendorCountry`` field - now add an additional split using the ``by`` clause. 

- With the ``by`` clause, we can further split our ``count`` by the values found in the ``product_name`` field (this results in a split of our initial ``count`` into separate columns in our statistical table):

This is also reflected in our visualization as well - can see here in the resulting visualization that we are keeping ``VendorCountry`` as our X-axis, but now the counts for Canada and the United States are further split by the count of each product. 

The values for a second split are represented by individually colored columns.

A legend for these values can be seen on the right of the visualization by default.

**By default, Splunk will display individual columns for the top 10 values found in the field you use to execute your Multi-series split.**

In this search, this would be the ``product_name`` field.

- By adding the argument of ``limit``, you can override that top 10 to whatever whole value you find appropriate:

Can see that by adding ``limit=5`` to our search string, it's resulted in only having the top five values for ``product_name`` appear as their own individual column. Any values less frequent than the top five get aggregated into the ``OTHER`` column. 

**Note that the ``limit`` argument sets a limit across the entire dataset - does not give you the top five in Canada and then a separate top five in the United States, for example.**

- Want to remove any limit on the number of columns shown, use ``limit=0``. 

Splunk will then give every possible value for your **Multi-series** split field its own column. 

If the column split has too many values to display in the visualization's legend, the list will include a down arrow to scroll through the values. 

- Can further control the column shown with the ``useother`` and ``usenull`` arguments (the ``useother`` argument will completely remove the ``OTHER`` column from your results):

Note that this is only a visual change - does not recalculate the original results that were generated by your base search.

The ``usenull`` argument will remove the ``NULL`` column if one exists. 

The ``NULL`` column will contain the statistical valuation of any events that did not actually contain the field that the **Multi-series** split was based on. 

In the search above, a ``NULL`` column would have been present if any of the events in the initial search returned did not contain a value for the ``product_name`` field.

> Additional examples of the ``chart`` command

Can see here that our count of events returned by our base search has been subjected to a **Single Series** split - have split that ``count`` by the values found in the ``vendor_action`` field. 

Here we see a **Multi-series** split - have first counted our events by ``vendor_action`` and then have further refined our table to split those actions by the different users (``user``) who took them. 

- Note: In this example, we are not using the ``over`` and the ``by`` clause, but simply a single instance of the by clause itself.

The last report was skewed by surplus categories - by using the ``useother=f`` argument, we have been able to remove these categories from our report.

The ``span`` argument allows for arbitrary grouping of values on the X-axis, as long as the X-axis is numeric.

This can group events into buckets - Splunk shifts overlapped values to the higher grouping - for example, a status code of 400 would have been bucketed in the 400 to 500 range, as opposed to the 300 to 400. 
 
The IT manager wants to see the volume of web activity over the last 24 hours by ranges of status codes. Without the ``span`` argument in place, each possible HTTP status code would have been given its own space on the X-axis, possibly resulting in an overly busy chart.
